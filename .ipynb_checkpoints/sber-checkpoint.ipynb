# Data Preparation

## Join DF

### Join

```python
combined_df = pd.merge(sessions_df, hits_df[['session_id', 'event_action']], on='session_id', how='left')
```

```python
combined_df.head()
```

### creating a key value

```python
combined_df = combined_df.drop_duplicates()
combined_df.shape
```

```python
def create_target_column(value):
    if value in ['sub_car_claim_click', 'sub_car_claim_submit_click',
'sub_open_dialog_click', 'sub_custom_question_submit_click',
'sub_call_number_click', 'sub_callback_submit_click', 'sub_submit_success',
'sub_car_request_submit_click']:
        return 1
    else:
        return 0
```

```python
combined_df['target_action'] = combined_df['event_action'].apply(create_target_column)
```

```python
combined_df = combined_df.drop('event_action', axis = 1)
```

```python
combined_df.target_action.value_counts()
```

```python
hits_df = None
sessions_df = None
```

## Data cleaning

```python
combined_df.head()
```

```python
combined_df.shape
```

```python
combined_df.isna().sum()
```

```python
combined_df.device_os = combined_df.device_os.apply(lambda x: '(not set)' if pd.isna(x) else x)
```

```python
combined_df.device_brand = combined_df.device_brand.apply(lambda x: '(not set)' if pd.isna(x) else x)
```

```python
for i in ['utm_source', 'utm_campaign', 'utm_adcontent', 'device_model', 'geo_city', 'geo_country']:
    combined_df[i] = combined_df[i].apply(lambda x: str(x).lower() )
```

```python
for i in ['utm_source', 'utm_campaign', 'utm_adcontent','device_model']:
    moda_value = combined_df[i].value_counts().idxmax()
    combined_df[i] = combined_df[i].apply(lambda x: moda_value if pd.isna(x) else x)
```

```python
combined_df.device_screen_resolution = combined_df.device_screen_resolution.apply(lambda x: 0.0 if x == '(not set)' else float(x.replace('x','.' )))
```

```python
def replace_strings_with_digits(strings):
    return str("(not set)" if re.search(r'\d', strings) else strings )
```

```python
combined_df.geo_country = combined_df.geo_country.apply(lambda x: replace_strings_with_digits(x))
```

```python
ohe = OneHotEncoder(sparse_output=False)
```

```python
for i in ['utm_source','utm_medium','utm_campaign','utm_adcontent','device_category','device_os','device_brand',
 'device_model','device_browser','geo_country']:
    ohe.fit(combined_df[[i]])
    ohe_i = ohe.transform(combined_df[[i]])
    combined_df[ohe.get_feature_names_out()] = ohe_i
```
